```python
import pandas as pd
```

```python
df = pd.read_csv('worldometer_data.csv')
```

```python
df
```

```python
df.head()
```

```python
Cases = df.TotalCases
```

```python
Cases
```

```python
Deaths = df.TotalDeaths
```

```python
Deaths
```

```python
from matplotlib import pyplot as plt
```

```python
plt.plot(Cases,Deaths, marker = 'd',color = 'g')
plt.xlabel('Cases')
plt.ylabel('Deaths')
plt.title('Cases vs Deaths')
plt.show()
```

```python
NewCases = df['NewCases'].sum() 
```

```python
NewDeaths = df['NewDeaths'].sum()
```

```python
plt.barh(['New Cases', 'New Deaths'], [NewCases, NewDeaths], color=['black', 'red'])
plt.xlabel('Count')
plt.ylabel('Category')
plt.title('COVID-19: New Cases and Deaths')
plt.show()
```

```python
Cases = df['TotalCases'].sum()
```

```python
Deaths = df['TotalDeaths'].sum()
```

```python
plt.pie([Cases, Deaths], labels = ['Cases','Deaths'], autopct = '%.2f%%')
plt.show()
```
