```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
a = np.array([1,2,3,4,5,6,7,8])
```

```python
b = a ** 2
```

```python
a
```

```python
b
```

```python
plt.plot(a,b, marker = 'd',color = 'g')
plt.xlabel('a value')
plt.ylabel('b value')
plt.title('A vs B')
plt.show()
```

```python
companies = ['apple', 'fb', 'google']
```

```python
revenue = [100,200,300]
```

```python
plt.bar(companies, revenue, color = 'g')
plt.xlabel('companies')
plt.ylabel('revenue')
plt.title('Company and Revenue')
plt.grid()
plt.show()
```

```python
plt.barh(companies, revenue, color = 'g')
plt.xlabel('companies')
plt.ylabel('revenue')
plt.title('Company and Revenue')
plt.grid()
plt.show()
```

```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df
```

```python
df.head()
```

```python
male = (df.Sex == 'male').sum()
```

```python
male
```

```python
female = (df.Sex == 'female').sum()
```

```python
female
```

```python
plt.bar(['male','female'], [male,female], color = 'g')
plt.xlabel('gender')
plt.ylabel('count')
plt.title('gender count')
plt.show()
```

```python
plt.barh(['male','female'], [male,female], color = 'g')
plt.xlabel('gender')
plt.ylabel('count')
plt.title('gender count')
plt.show()
```

```python
plt.pie([male, female], labels = ['male','female'], autopct = '%.2f%%')
plt.show()
```
